## LASSO特征选择

```python
## LASSO
Lambdas = np.arange(0.0001,0.2,0.005)
# 构造空列表，用于存储模型的偏回归系数
lasso_cofficients = []
for Lambda in Lambdas:
    lasso = Lasso(alpha=Lambda, normalize=True, max_iter=10000)
    lasso.fit(data_x,data[['group']])  ##     .....xtr,ytr.iloc[:,0]
    lasso_cofficients.append(lasso.coef_)
'''
可视化方法确定λ的值
'''
# 绘制Lambda与回归线的折线图
plt.plot(Lambdas, lasso_cofficients)
# 对x轴做对数变换
plt.xscale('log')
# 设置折线图x轴和y轴标签
plt.xlabel('Lambda')
plt.ylabel('Cofficients')
# 显示图形
plt.savefig('lasso.pdf',dpi=300,bbox_inches='tight')
plt.show()


lasso = Lasso(alpha=0.1,random_state=2021)  ## 
lasso.fit(data_x,data[['group']])

xtr_lasso = data_x.iloc[:,abs(lasso.coef_)>0]
# xte_lasso = xte.iloc[:,abs(lasso.coef_)>0]


xtr_lasso.columns
```

```python


# lasso选择,定义子数据集。
lr_col_2 = xtr_lasso.columns.tolist()
da_X = data[lr_col_2]
da_Y = data[["group"]]
xte = xte[lr_col_2]


xte_rscs1 = xte_rscs1[lr_col_2]
xte_rscs2 = xte_rscs2[lr_col_2]
xte_num1 = xte_num1[lr_col_2]
xte_num2 = xte_num2[lr_col_2]
xte_num3 = xte_num3[lr_col_2]
```

# 建模

```python
def plot_auc(y_test,y_score,label,title,picname):
    plt.rc('font',family='Times New Roman')
    fpr,tpr,threshold = roc_curve(y_test, y_score) ###计算真正率和假正率
    roc_auc = auc(fpr,tpr) ###计算auc的值
    title=title
    picname=picname
    label=label
    lw = 2
    plt.figure(figsize=(5,5))
    plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='AUC {}'.format(label)) ###假正率为横坐标，真正率为纵坐标做曲线
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characterist {}'.format(title))
    plt.legend(loc="lower right")
#     plt.show()
    plt.savefig(picname)
def plot_cali(y_test,y_score,clf_names,title,picname):
    plt.rc('font',family='Times New Roman')
    title = title
    picname = picname
    skp.metrics.plot_calibration_curve(y_test,y_score,clf_names,title=title,figsize=(5,5),n_bins=3,cmap='rainbow')
    plt.savefig(picname)
```

## 随机森林

```python
from sklearn.pipeline import make_pipeline
kv=KFold(n_splits=10,shuffle=True,random_state=0)
pipeline = make_pipeline(RandomForestClassifier(n_estimators=100, max_depth=4,random_state=0))
scores = []
for k, (train, test) in enumerate(kv.split(da_X,da_Y)):
#     print(k)
    pipeline.fit(da_X.iloc[train, :], da_Y.iloc[train])
    score = pipeline.score(da_X.iloc[test, :], da_Y.iloc[test])
    scores.append(score)
scores
```

```python
for k, (train, test) in enumerate(kv.split(da_X,da_Y)):
    if k==9:
        trainDataX = da_X.iloc[train, :]
        trainDataY =da_Y.iloc[train]
        valDataX=da_X.iloc[test, :]
        valDataY=da_Y.iloc[test]
```

```python
xtr=trainDataX
ytr=trainDataY
```

```python
rf = RandomForestClassifier(max_depth =100,max_features=5,max_leaf_nodes=20,min_samples_split=50,n_estimators=50,random_state=0)
rf.fit(xtr,ytr)

# rf =LogisticRegression()
# rf.fit(xtr,ytr)
```

```python
data_train_p = mm.model_parameter(ytr.iloc[:,0],rf.predict_proba(xtr)[:,1],auc_ci_type=1,n_bins=5)
data_train_p
```

```python
data_test_p = mm.model_parameter(yte.iloc[:,0],rf.predict_proba(xte)[:,1],auc_ci_type=1,n_bins=3,cut_off=data_train_p['Cut_off'].astype(np.float).values)
data_test_p
```
